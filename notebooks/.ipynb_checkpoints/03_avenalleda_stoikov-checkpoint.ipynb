```python
## Class Avenalleda-Stoikov

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

class AvellanedaStoikovAgent:
    
    def __init__(self, gamma, kappa, T, dt=1):
        self.gamma = gamma
        self.kappa = kappa
        self.T = T
        self.dt = dt
        self.q = 0
        self.time = 0

    def reservation_price(self, mid_price, volatility):
        time_left = self.T - self.time
        r = mid_price - (self.q * self.gamma * volatility**2 * time_left)
        return r

    def optimal_spread(self, volatility):
        time_left = self.T - self.time
        opt_spread = 1/self.gamma * np.log(1 + self.gamma/self.kappa) + 0.5 * self.gamma * volatility**2 * time_left
        return opt_spread

    def get_quotes(self, mid_price, volatility):
        r = self.reservation_price(mid_price, volatility)
        delta = self.optimal_spread(volatility)

        bid = r - delta
        ask = r + delta
        return bid, ask

    def update_inventory(self, fill_side, fill_size):
        if fill_side == 'buy':
            self.q += fill_size
        if fill_side == 'sell':
            self.q -= fill_size

    def step_time(self):
        self.time += self.dt

## Integration Kalman + Avenalleda-Stoikov

df = pd.read_csv('../data/strategy_results.csv', index_col=0, parse_dates=True)
strategy_df = df.copy()

strategy_df['volatility'] = strategy_df['spread_return'].rolling(20).std()
strategy_df = strategy_df.iloc[20:].copy()

bids = []
asks = []
inventories = []
reservation_prices = []

mm_agent = AvenalledaStoikovAgent(gamma=0.1, kappa=1.5, T=len(strategy_df))

strategy_df['position_change'] = strategy_df['position'].diff()

for index, row in strategy_df.iterrows():
    mid_price = row['spread_kalman']
    bid, ask = mm_agent.get_quotes(mid_price=mid_price, volatility=row['volatility'])
    
    bids.append(bid)
    asks.append(ask)

    r = mm_agent.reservation_price(mid_price, row['volatility'])

    reservation_prices.append(r)
    inventories.append(mm_agent.q)

    if row['position_change'] > 0:
        mm_agent.update_inventory('buy', 1)
    elif row['position_change'] < 0:
        mm_agent.update_inventory('sell', 1)
        
    mm_agent.step_time()
    
strategy_df['bids'] = bids
strategy_df['asks'] = asks
strategy_df['inventories'] = inventories
strategy_df['reservation_prices'] = reservation_prices

## AS Visualization

fig, axes = plt.subplots(3, 1, figsize=(12,12))

axes[0].plot(strategy_df.index, strategy_df['spread_kalman'], label='Spread Central (KALMAN)', color='black')
axes[0].plot(strategy_df.index, strategy_df['bids'], color='green', label='Bid')
axes[0].plot(strategy_df.index, strategy_df['asks'], color='red', label='Ask')
axes[0].set_title('AS Market-Making: Spread and Bid/Ask Quotes')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

axes[1].plot(strategy_df.index, strategy_df['spread_kalman'], label='Spread Central (KALMAN)', color='black')
axes[1].plot(strategy_df.index, strategy_df['reservation_prices'], label='Reservation Price', color='purple')
axes[1].set_title('Reservation Price vs Kalman')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

axes[2].plot(strategy_df.index, strategy_df['inventories'], label= 'Inventories', color='blue')
axes[2].set_title('Inventory Over Time')
axes[2].axhline(0, color='black', linestyle='--', linewidth=1, alpha=0.5)
axes[2].legend()
axes[2].grid(True,alpha=0.3)

plt.tight_layout()
plt.savefig('../results/03_avenalleda_stoikov.png', dpi=150)
plt.show()
```
