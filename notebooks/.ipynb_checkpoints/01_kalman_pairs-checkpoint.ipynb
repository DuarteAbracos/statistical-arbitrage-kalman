```python
import pandas as pd
import yfinance as yf
from statsmodels.tsa.stattools import coint
import requests
import itertools
from datetime import datetime, timedelta
import io
import os
from alpaca.data.historical import StockHistoricalDataClient
from alpaca.data.requests import StockBarsRequest
from alpaca.data.timeframe import TimeFrame
```

```python
## Kalman Filter for Pairs Trading 

#**Objective:** Estimate time-varying hedge ratio β(t) for pairs trading using Kalman Filtering.

#**Approach:**
#1. Systematic pair selection across S&P 100 by sector
#2. Cointegration testing (Augmented Dickey-Fuller)
#3. Kalman Filter with EM algorithm for Q/R calibration
#4. Comparison with static OLS baseline


# ============================================
# 1. DATE SETUP (CORRECTED FOR TRAIN/TEST SPLIT)
# ============================================
start_date = '2020-01-01'
train_end_date = '2023-12-31' 

yesterday = datetime.now() - timedelta(days=1)
end_date = yesterday.strftime('%Y-%m-%d')

headers = {"User-Agent": "Mozilla/5.0"}
wiki_url = 'https://en.wikipedia.org/wiki/S%26P_100'
response = requests.get(wiki_url, headers=headers)
sp100_data = pd.read_html(io.StringIO(response.text))
sp100_table = sp100_data[2]

indices_sectors = ['SPY', 'QQQ', 'DIA', 'IWM', 'EEM', 'VGK', 'EWJ', 'XLF', 'XLE', 'XLK', 'XLV', 'XLY', 'XLI', 'XLC', 'XLU', 'XLB', 'XLP', 'SMH', 'GDX', 'SLV', 'IBB', 'XRT', 'VTI', 'VEA', 'VWO']

all_categories = {}

for sector, group in sp100_table.groupby('Sector'):
    symbols = group['Symbol'].str.replace('.', '-').tolist()
    all_categories[f"Stocks - {sector}"] = symbols

all_categories["Indices"] = indices_sectors

overall_best_pair = None
overall_min_p = 1.0

print("Searching for the best pairs in each category (using ONLY Train Data)...")

for category_name, ticker_group in all_categories.items():
    # Failsafe: Prevent mathematical errors if a sector has fewer than 2 assets
    if len(ticker_group) < 2:
        continue
        
    print(f"\nProcessing category: {category_name}...")
    
    volume_check = yf.download(ticker_group, start=start_date, end=end_date, progress=False)['Volume']
    liquid_list = volume_check.mean().sort_values(ascending=False).head(25).index.tolist()

    category_data = yf.download(liquid_list, start=start_date, end=end_date, progress=False)['Close']
    
    category_pairs = list(itertools.combinations(liquid_list, 2))
    best_pair_in_cat = None
    min_p_in_cat = 1.0
    
    for tick1, tick2 in category_pairs:
        if tick1 in category_data.columns and tick2 in category_data.columns:

            pair_prices = category_data[[tick1, tick2]].dropna()

            if pair_prices.empty:
                continue
            
            # ============================================
            # 2. THE GRAND FIX: SLICING DATA FOR BLIND TESTING
            # ============================================
            pair_prices_train = pair_prices.loc[:train_end_date]
            
            # Ensure there is enough historical data in the training set
            if len(pair_prices_train) < 100: 
                continue
            
            p_X_train = pair_prices_train[tick1]
            p_Y_train = pair_prices_train[tick2]
    
            # Evaluate cointegration STRICTLY on past data (Training Set)
            _, p_value, _ = coint(p_X_train, p_Y_train)

            if p_value > 0.05:
                continue
            
            if p_value < min_p_in_cat:
                min_p_in_cat = p_value
                best_pair_in_cat = (tick1, tick2)
                
                if p_value < overall_min_p:
                    overall_min_p = p_value
                    overall_best_pair = (tick1, tick2)
                    
                    # We save the full series (Train + Test) for the backtester later, 
                    # but the pair was CHOSEN using ONLY Train data!
                    prices_X = pair_prices[tick1] 
                    prices_Y = pair_prices[tick2]
                    
    if best_pair_in_cat is None:
        print(f"Category {category_name} had no valid pair.")
    else: 
        print(f"Category {category_name} winner: {best_pair_in_cat} with p-value {min_p_in_cat:.4f}")

print(f"\n🏆 THE BEST PAIR IS: {overall_best_pair} with p-value {overall_min_p:.6f}")

print(f"\n Downloading data from Alpaca to {overall_best_pair}...")
print("Might take a while, there is a lot of data.")

API_KEY = 'PKUWS7A35BIY6Q3ZUL3RYZ2BZO'
API_SECRET = '2S3ALiAzcZ29jdpPMz3LzHzYDxd6LiwM71oXaxhVuzzo'

client = StockHistoricalDataClient(API_KEY, API_SECRET)

request_params = StockBarsRequest(
    symbol_or_symbols=list(overall_best_pair),
    timeframe=TimeFrame.Minute,
    start=datetime.strptime(start_date, '%Y-%m-%d'),
    end=datetime.strptime(end_date, '%Y-%m-%d')
)

alpaca_bars = client.get_stock_bars(request_params).df
alpaca_bars = alpaca_bars.reset_index()

pivot_df = alpaca_bars.pivot(index='timestamp', columns='symbol')

ticker_X = overall_best_pair[0]
ticker_Y = overall_best_pair[1]

df_intra = pd.DataFrame({
    'X_close': pivot_df['close'][ticker_X],
    'Y_close': pivot_df['close'][ticker_Y],
    'Y_high': pivot_df['high'][ticker_Y],
    'Y_low': pivot_df['low'][ticker_Y]
}).dropna()

print(f"Success! We have {len(df_intra)} bars perfectly aligned.")
```

```python
from sklearn.linear_model import LinearRegression

pair_df = pd.DataFrame({chosen_pair[0] : prices_X, chosen_pair[1] : prices_Y}).dropna()

pairtick0 = pair_df[chosen_pair[0]].values.reshape(-1, 1)
pairtick1 = pair_df[chosen_pair[1]].values

model_ols = LinearRegression()
model_ols.fit(pairtick0, pairtick1)
beta_ols = model_ols.coef_[0]
alpha_ols = model_ols.intercept_
print(f"OLS Static: α={alpha_ols:.4f}, β={beta_ols:.4f}")
spread_ols = pairtick1 - (model_ols.intercept_ + (beta_ols * pairtick0.flatten()))
```

```python
## Kalman Filter Implementation

#State vector: x_t = [α_t, β_t]  
#Observation: Y_t = α_t + β_t · X_t + noise  
#Transition: Random walk (F = I)

from pykalman import KalmanFilter
import numpy as np

obs_matrices = np.array([[1, x] for x in pairtick0.flatten()]).reshape(-1, 1, 2)

kf = KalmanFilter(
    n_dim_obs=1,
    n_dim_state=2,
    initial_state_mean=[0, 1],
    observation_matrices=obs_matrices,
    transition_matrices = np.eye(2),
    observation_covariance=1,
    transition_covariance=np.eye(2) * 0.01,
)
kf = kf.em(pairtick1, n_iter=10, em_vars=['transition_covariance', 'observation_covariance'])

state_means, state_covs = kf.filter(pairtick1)

alpha_kalman = state_means[:, 0]
beta_kalman = state_means[:, 1]

spread_kalman = pairtick1 - (alpha_kalman + (beta_kalman * pairtick0.flatten()))
```

```python
import matplotlib.pyplot as plt

def zscore(spread):
    return((spread - spread.mean()) / spread.std())
z_ols = zscore(pd.Series(spread_ols))
z_kalman = zscore(pd.Series(spread_kalman))

fig, axes = plt.subplots(3, 1, figsize=(12, 10))
fig.suptitle(f"THE BEST PAIR IS: {overall_best_pair} with p-value {overall_min_p:.6f}\n", fontsize=18)

dates = pair_df.index

axes[0].plot(dates, beta_kalman, label='Dynamic Kalman β', color='blue', alpha=0.7)
axes[0].plot(dates, beta_ols * np.ones(len(beta_kalman)), label='Static OLS β', color='red')
axes[0].set_title('Hedge Ratio: OLS vs Kalman')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

axes[1].plot(dates, spread_kalman, label='Kalman Spread', color='blue', alpha=0.5)
axes[1].plot(dates, spread_ols, label='OLS Spread', color='red', alpha=0.5)
axes[1].axhline(0, color='black', linestyle='--', linewidth=1)
axes[1].set_title('Spread Comparison')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

axes[2].plot(dates, z_ols, label='Z-score (OLS)', alpha=0.5)
axes[2].plot(dates, z_kalman, label='Z-score (Kalman)', alpha=0.5)
axes[2].axhline(2, color='red', linestyle='--', alpha=0.5)
axes[2].axhline(-2, color='green', linestyle='--', alpha=0.5)
axes[2].set_title('Z-Score of Spread')
axes[2].legend()
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('../results/01_kalman_comparison.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
## Quantitative Analysis

#Beta statistics
#Spread statistics
#Z-score analysis
#Improvement metric

print("\n" + "="*50)
print("KALMAN FILTER PERFORMANCE")
print("="*50)

print(f"\nBeta (OLS Static):     {beta_ols:.4f}")
print(f"Beta (Kalman Mean):    {beta_kalman.mean():.4f}")
print(f"Beta (Kalman Std):     {beta_kalman.std():.4f}")
print(f"Beta (Kalman Min):     {beta_kalman.min():.4f}")
print(f"Beta (Kalman Max):     {beta_kalman.max():.4f}")

print(f"\nSpread OLS Mean:       {spread_ols.mean():.4f}")
print(f"Spread OLS Std:        {spread_ols.std():.4f}")
print(f"Spread Kalman Mean:    {spread_kalman.mean():.4f}")
print(f"Spread Kalman Std:     {spread_kalman.std():.4f}")

print(f"\nZ-score OLS >2:        {(np.abs(z_ols) > 2).sum()} days ({(np.abs(z_ols) > 2).sum()/len(z_ols)*100:.1f}%)")
print(f"Z-score Kalman >2:     {(np.abs(z_kalman) > 2).sum()} days ({(np.abs(z_kalman) > 2).sum()/len(z_kalman)*100:.1f}%)")

spread_improvement = (1 - spread_kalman.std() / spread_ols.std()) * 100
print(f"\n Kalman reduced spread volatility by {spread_improvement:.1f}%")
print("="*50)
```

```python
results_df = pair_df.copy()
results_df['beta_ols'] = beta_ols
results_df['beta_kalman'] = beta_kalman
results_df['alpha_kalman'] = alpha_kalman
results_df['spread_ols'] = spread_ols
results_df['spread_kalman'] = spread_kalman
results_df['z_ols'] = z_ols
results_df['z_kalman'] = z_kalman.values

results_df.to_csv('../data/kalman_results.csv')

metadata = {
    'pair': overall_best_pair,
    'p_value': overall_min_p,
    'beta_ols': beta_ols,
    'start_date': start_date,
    'end_date': end_date
}

import json
with open('../data/metadata.json', 'w') as f:
    json.dump(metadata, f, indent=2)

print("Data saved to data/kalman_results.csv")
print("Metadata saved to data/metadata.json")
```
